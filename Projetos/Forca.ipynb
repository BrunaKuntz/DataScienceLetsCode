```python
# Importa√ß√µes
from random import randint
from time import sleep
from google.colab import output
import pandas as pd
import emoji

# Boas vindas anima√ß√£o
for i in range(2):
  print(r''' ______________________________________           üòä
|     BEM-VINDO AO JOGO DA FORCA       |          /|\
|    ---------------------------       |         / | \ 
|     Se prepare para o desafio        |          / \
 ______________________________________          /   \
''')
  sleep(1.7)
  output.clear()
  print(r''' ______________________________________          \üòâ /
|     BEM-VINDO AO JOGO DA FORCA       |          \|/
|    ---------------------------       |           | 
|     Se prepare para o desafio        |          / \
 ______________________________________          /   \
''')
  sleep(1.7)
  output.clear()


# nome jogador
nomejogador = input(r''' ______________________________________           üòä
|     BEM-VINDO AO JOGO DA FORCA       |          /|\
|    ---------------------------       |         / | \ 
|     Qual √© o nome do jogador?        |          / \
 ______________________________________          /   \
        ''')
output.clear()

# Banco de Palavras
df = pd.read_csv('tabela palavras.csv')

# Randomiza√ß√£o da categoria
n = randint(0, 6)

if n == 0:
  categoria = 'Animal'
elif n == 1:
  categoria = 'Cor'
elif n == 2:
  categoria = 'Lugares'
elif n == 3:
  categoria = 'Comida'
elif n == 4:
  categoria = 'Profiss√µes'
elif n == 5:
  categoria = 'Minha sogra √©'
else:
  categoria = 'Objeto'

# Randomiza√ß√£o da palavra
palavra = df[categoria][randint(0, 48)]
letraspalavra = []

for i in range(len(palavra)):
  letraspalavra.append(palavra[i])
  
vidas = 5

print(palavra)

# Desenho Forca 
forca = [r"""_______           
|
|
|
|
|
|   
|
""", 
r"""_______           
|
|     üòÅ
|    
|  
|    
|   
|
""", 
r"""_______           
|
|    ü§®
|    /|\
|   
| 
|  
|
""", 
r"""_______           
|
|    üòÆ
|    /|\
|   / | \
|  
| 
| 
""", 
r"""_______           
|
|    üò≤
|    /|\
|   / | \
|    / \
|   
|
""",
r"""_______           
|
|    üíÄ
|    /|\
|   / | \
|    / \
|   /   \
|
"""] 

# jogo
c = 0 # quantidade de acertos
listaletras = ""
f = 0 # desenho forca
chutes = ['_']*len(letraspalavra)

while vidas != 0:
  coracao = '‚ù§'*(vidas)

  # menu
  print(f'Jogador: {nomejogador}                       Categoria: {categoria}')
  print(f'Vidas: {coracao}                         Letras chutadas:{listaletras}')

  # desenho forca
  print(forca[f])
  
  print(' '.join(chutes))
  # palpite
  palpite = input('Chute uma letra: ')

  while palpite in listaletras:
    palpite = input('Voc√™ j√° deu esse palpite. Tente um novo: ')

  listaletras = listaletras + palpite

  if palpite in letraspalavra:
    for i in range(len(letraspalavra)):
      if palpite == letraspalavra[i]:
        chutes[i] = letraspalavra[i]
        c += 1
  else:
    f +=1
    vidas -= 1

  output.clear()

  if c == (len(letraspalavra)):

     # menu
    print(f'Jogador: {nomejogador}                       Categoria: {categoria}')
    print(f'Vidas: {coracao}                         Letras chutadas:{listaletras}')

    # desenho forca
    print(forca[f])
    print(' '.join(chutes))
    textofinal = f'VOC√ä ACERTOU! A palavra era {palavra}'
    break 

if vidas == 0:
  # menu
  print(f'Jogador: {nomejogador}                       Categoria: {categoria}')
  print(f'Vidas: ‚ùå                         Letras chutadas:{listaletras}')

  # desenho forca
  print(forca[f])
  print(' '.join(chutes))
  textofinal = f'GAME OVER! Voc√™ n√£o conseguiu descobrir a palavra {palavra}' 

print(f'{textofinal}\n')
```
